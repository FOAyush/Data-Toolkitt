**2) How does broadcasting work in NumPy?**

**Ans->** Broadcasting in NumPy is a powerful feature that allows arrays of different shapes to be used together in arithmetic operations without explicitly reshaping or replicating data. When performing operations on two arrays, NumPy compares their shapes element-wise, starting from the trailing dimensions. If the dimensions are equal or if one of them is 1, they are considered compatible. In cases where one array has fewer dimensions or a dimension of size 1, NumPy "stretches" or "broadcasts" it across the larger array so the operation can proceed element-wise. This enables efficient computation without the need for manual looping or creating large intermediate arrays, making code more concise and faster. For example, adding a 1D array to a 2D array will result in the 1D array being broadcast across each row or column, depending on its shape. Broadcasting is a key feature that underpins many high-performance operations in NumPy.

6)What are the differences between NumPy arrays and Python lists

Ans-> NumPy arrays and Python lists both serve as containers for storing collections of elements, but they differ in several key aspects. One major difference is performance—NumPy arrays are significantly faster and more memory-efficient than Python lists, particularly when dealing with large datasets. Additionally, NumPy arrays offer a wide range of built-in functions and capabilities for mathematical, statistical, and linear algebra operations that are not available with native Python lists. These differences make NumPy arrays a preferred choice for scientific computing and data analysis tasks where performance and numerical capabilities are essential.

**8) What does the term “vectorized operation” mean in NumPy?**

Ans-> In NumPy, a **vectorized operation** refers to performing element-wise computations on entire arrays or large blocks of data at once, without using explicit Python loops. Instead of iterating through each element one by one, NumPy applies operations directly to whole arrays, leveraging optimized, low-level implementations in C for faster execution. This not only makes the code more concise and readable but also significantly improves performance by taking advantage of hardware-level optimizations and parallelism. For example, adding two arrays of the same shape using a vectorized operation simply adds corresponding elements together in one step, rather than looping through each element manually. Vectorized operations are a core feature of NumPy that enable efficient numerical computing on large datasets.

**17) Explain the difference between apply() and map() in Pandas.**

**Ans->** In Pandas, both `apply()` and `map()` are used to **apply functions** to data, but they differ in their scope and typical use cases.

`map()` is primarily used on **Series** objects to **map or substitute each value** in the Series according to a function, dictionary, or another Series. It is best suited for element-wise transformations or value replacements, such as mapping categorical values to numbers or applying a simple function to each element.

On the other hand, `apply()` is more flexible and can be used on both **Series and DataFrames**. When used on a Series, it applies a function to each element, similar to `map()`, but can handle more complex operations. When used on a DataFrame, `apply()` can apply a function along either rows or columns, allowing for row-wise or column-wise computations, aggregations, or transformations.

In summary, use `map()` for straightforward element-wise mapping on a Series, and use `apply()` when you need more flexibility, such as applying functions across rows or columns in a DataFrame or performing more complex operations on Series.

**21) Why is NumPy’s array slicing faster than Python’s list slicing?**

**Ans->** NumPy’s array slicing is faster than Python’s list slicing primarily because of how the data is stored and accessed. NumPy arrays are stored as **contiguous blocks of memory** with elements of the same data type, which allows slicing operations to create **views** on the original data without copying it. This means when you slice a NumPy array, it simply creates a new reference to the existing data, making the operation very fast and memory-efficient. In contrast, Python lists store references to objects scattered in memory and can contain elements of different types, so slicing a list involves creating a **new list** and copying references to the sliced elements, which is slower and uses more memory. Additionally, NumPy’s slicing takes advantage of optimized, low-level C implementations, while Python list operations rely on higher-level Python code. These factors combined make NumPy slicing much faster and more efficient than list slicing.

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)


corr_matrix = df.corr()


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')


plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#13 Create a 3D scatter plot using Plotly

import plotly.graph_objs as go
import numpy as np


np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
z = np.random.randn(50)


fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])

fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    title='3D Scatter Plot Example'
)

fig.show()
```
